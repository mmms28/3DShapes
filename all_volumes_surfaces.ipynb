## Define global image parameters

```python
out_dir = "Examples_Volumes"
image_dimensions = (256, 256, 256)  # Image dimensions
image_spacing = (1.0, 1.0, 1.0)  # Voxel spacing
image_origin = (0, 0, 0)

center = (image_dimensions[0] // 2, image_dimensions[1] // 2, image_dimensions[2] // 2)  # Define center of the shape

print(f"{'-'*10} Image {'-'*10}\n Dimentions: {image_dimensions} \n Spacing: {image_spacing} \n Origin: {image_origin} \n Object center: {center} ")
```

```python
def save_and_plot_volume_and_object(image: vtk.vtkImageData, file_name: str, out_dir: str = "Examples", verbose: bool=False):
    
    # Save the vtkImage object as a .mhd file.
    write_vtkimage(image, f"{out_dir}/volumes/{file_name}.mhd", verbose=verbose)
    
    # Create a vtk Polydata from the image.
    polydata =  create_polydata_from_vtkImage(image)

    # Save polydata object as .vtk file
    write_object(polydata, f"{out_dir}/objects/{file_name}.vtk", verbose=verbose)

    # Plot volume
    plot_views(image=image, image_name=file_name, display=verbose, save=True, output_dir=f"{out_dir}/images/volumes")

    # Plot object
    plot_vtkPolyData(polydata=polydata, polydata_name=file_name, save=True, output_dir = f"{out_dir}/images/objects") 
    
```

## Create, save and display shapes

### Sphere

When $ a = b = c $ , the ellipsoid is a sphere

```python
radii = (40,40,40)  # Sphere radius

#Create sphere
sphere_image, sphere_name = create_ellipsoid_volume(radii=radii, dimensions=image_dimensions, center=center, 
                                                    spacing=image_spacing, origin=image_origin)
save_and_plot_volume_and_object(image=sphere_image, file_name=sphere_name, out_dir=out_dir, verbose=True)
```
