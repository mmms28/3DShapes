# Create and display shapes using utils

## Import libraries

```python
import os, vtk
from create_3D_utils import *
from manipulate_3D_utils import *
```

## Define global image parameters

```python
out_dir = "Examples_Parametric"
image_dimensions = (256, 256, 256)  # Image dimensions
image_spacing = (1.0, 1.0, 1.0)  # Voxel spacing
image_origin = (0, 0, 0)

center = (image_dimensions[0] // 2, image_dimensions[1] // 2, image_dimensions[2] // 2)  # Define center

print(f"{'-'*10} Image {'-'*10}\n Dimentions: {image_dimensions} \n Spacing: {image_spacing} \n Origin: {image_origin} \n Object center: {center} ")
```

## Define functions

```python
def save_and_plot_object_and_volume(polydata: vtk.vtkPolyData, file_name: str, out_dir: str = "Examples", verbose: bool=False):
    # Save the vtkPolyData object as a .vtk file.
    write_object(polydata, f"{out_dir}/objects/{file_name}.vtk", verbose=verbose)
    
    # Create a volume image from the polydata.
    image =  create_vtkimage_from_polydata(dimensions=image_dimensions, spacing=image_spacing,
                                                    origin=image_origin, vtk_polydata=polydata)
    # Save volume image as .mhd file
    write_vtkimage(image, f"{out_dir}/volumes/{file_name}.mhd", verbose=verbose)
    
    # Plot object
    plot_vtkPolyData(polydata=polydata, polydata_name=file_name, save=True, output_dir = f"{out_dir}/images/objects") 
    
    # Plot volume
    plot_views(image=image, image_name=file_name, display=verbose, save=True, output_dir=f"{out_dir}/images/volumes")
```

## Create, save and display shapes

#### Sphere
When $ a = b = c $ , the ellipsoid is a sphere

```python
radii=[100,100,100]
resolution = 100 

sphere_polydata, sphere_name = create_ellipsoid_object(resolution=resolution, radii=radii, center=center)
save_and_plot_object_and_volume(polydata=sphere_polydata, file_name=sphere_name, out_dir=out_dir, verbose=True)
```
